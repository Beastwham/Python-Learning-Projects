```python
"""
Question: Given a positive integer 'n' less than or equal to 26, you are required to print the below pattern 
 
Sample Input: 5 
 
Sample Output : 
--------e-------- 
------e-d-e------ 
----e-d-c-d-e---- 
--e-d-c-b-c-d-e-- 
e-d-c-b-a-b-c-d-e 
--e-d-c-b-c-d-e-- 
----e-d-c-d-e---- 
------e-d-e------ 
--------e-------- 
 
Sample Input  : 3 
 
Sample Output : 
----c---- 
--c-b-c-- 
c-b-a-b-c 
--c-b-c-- 
----c----

"""

# Solution
import string
alphabet_list = list(string.ascii_lowercase)
n = int(input("Enter a Number: "))
l = alphabet_list[0:n]
for i in range(1, n+1): # Loop 1 for first 5 rows
    for j in range(0, n-i): #First --
        print("--", end = "")
    
    for k in range(1,i+1):# Then first sequence of letters - runs in first sequence as well
        if(k==i):
            print(l[-k], end = "")
        else:
            print(l[-k], end = "-")
    
    for x in range(1,i): #Second sequence of letters beyond the [-i] element
        print(f"-{l[-(i-x)]}", end = "")
    
    for y in range(0, n-i):
        print("--", end = "")
    
    
    print("")

    
for i in range(1,n): #loop 2 for last 4 rows

    ##print("--"*i, end = "") - can be used instead of for loop for --
    
    for j in range(0,i):
        print("--", end = "")
    
    for k in range(1, (n-i+1)):
        if(k==n-i):
            print(l[-k], end = "")
        else:
            print(l[-k], end = "-")
    
    for x in range(n-i, 1, -1):
        print(f"-{l[-(x-1)]}", end = "")
  
    ##print("--"*i, end = "")
    for j in range(0,i):
        print("--", end = "")
        
    print("")
```
